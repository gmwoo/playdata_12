   각 그룹의 최대값과 최소값의 차리를 계산하여 그룹별로 집계해 보세요.

5. fare 열에는 min, max 함수를, age 열에는 mean 함수를 적용해 보세요.
6. age 열의 평균이 30보다 작은 그룹만을 필터링하여 데이터프레임으로 반환하세요

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.simplefilter(action="ignore", category = FutureWarning)
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# 1번
df1 = df[['age','sex','class','fare','survived']]
df1
```

```python
df2 = df1.groupby('class').agg(np.std).reset_index()
df2
```

```python
def cal_min_max(x):
    return x.max() - x.min()
```

```python
df3 = df1.groupby('class').agg(cal_min_max)
df3
```

```python
df4 = df1.groupby('class').agg(['min', 'max'])
df4
```

```python
df5 = df1.groupby('class').agg({
    "fare":['min', 'max'],
    "age":['mean']})
df5
```

```python
df6 = df1.groupby('class').filter(lambda x: x['age'].mean() < 30)
df6.tail()
```
